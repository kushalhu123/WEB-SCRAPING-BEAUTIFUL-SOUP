```python
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup

response=requests.get('http://books.toscrape.com//catalogue/page-1.html')

html=response.text

data=BeautifulSoup(html,"html.parser")

alllinks=[]
alllinks.append("http://books.toscrape.com//catalogue/page-1.html")
baseurl="http://books.toscrape.com//catalogue/"
while response.status_code==200:
    nextt=data.find(class_="next")
    if nextt is None:
        break
    nextt=nextt.a['href']
    newurl=baseurl+nextt
    #print(newurl)
    alllinks.append(newurl)
    response=requests.get(newurl)
    html=response.text
    data=BeautifulSoup(html,"html.parser")
print(alllinks)

n=len(alllinks)

print(n)

totalbooks=[]
allbooklinks=[]
baseurl="http://books.toscrape.com//catalogue/"
for i in range(5):
    currenturl=alllinks[i]
    response=requests.get(currenturl)
    html=response.text
    data=BeautifulSoup(html,"html.parser")
    classs=data.find_all(class_="product_pod")
    for link in classs:
        allbooklinks.append(baseurl+link.h3.a['href'])
    
    for booklink in allbooklinks:
        response=requests.get(booklink)
        html=response.text
        data=BeautifulSoup(html,"html.parser")
        title=data.h1.string
        price=data.find(class_="price_color")
        price=price.string
        price=float(re.search('[\d.]+',price).group())
        stocks=data.find(class_="instock availability")
        stock=stocks.contents[-1].strip()
        stock=re.search('\d+',stock).group()
        istock=stocks.contents[-1].strip()
        isstock=' '.join(istock.split()[0:2])
        url=booklink
        rating=data.find('p',attrs={'class':'star-rating'}).get('class')[-1]
        wht=data.find_all('a')
        wht=wht[-1].string
        imgg=data.find('div',attrs={'class':'item active'})
        imgg=imgg.img['src']
        totalbooks.append([title,url,price,stock,isstock,rating,wht,imgg])

df=pd.DataFrame(totalbooks,columns=['title','url','price','stock','isstock','rating','what-type','img'])
df
```
